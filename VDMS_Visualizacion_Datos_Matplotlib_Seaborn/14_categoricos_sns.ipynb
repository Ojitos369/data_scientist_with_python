```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(data=tips, x="day", hue="sex")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.stripplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, color="black", palette='dark:black', marker="*")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, color="black", palette='dark:black', marker="*")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", palette='pastel', split=True)
# sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, marker="*")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(data=tips, y="total_bill", hue="sex", palette='pastel', split=True)
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, kind="swarm", col="smoker", row="time")
plt.show()
```
