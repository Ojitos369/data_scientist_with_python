```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
plt.style.available
```

```python
plt.style.use('dark_background')
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(x, x + 1, 'r:o')
ax.plot(x, x + 2, 'b-+')
ax.plot(x, x + 3, 'g-.d')
ax.plot(x, x + 4, 'y--.')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(-x, x + 1, color='red')
ax.plot(-x, x + 2, color='green')
ax.plot(-x, x + 3, color='blue')
ax.plot(-x, x + 4, color='yellow')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(x, x - 1, color='#bb0000', alpha=0.5)
ax.plot(x, x - 1.2, color='#a008')
ax.plot(x, x - 1.4, color='#a00')

ax.plot(x, x - 2, color='#00bb00', linewidth=3)
ax.plot(x, x - 2.2, color='#0a0')

ax.plot(x, x - 3, color='#0000bb', linestyle='--')
ax.plot(x, x - 3.2, color='#00a')

ax.plot(x, x - 4, color='#bbbb00', marker='o', markersize=3)
ax.plot(x, x - 4.2, color='#aa0', marker='x', markersize=10)
ax.plot(x, x - 4.4, color='#990', marker='D', markerfacecolor='#00bbbb')


plt.show()
```
