```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.histplot(data=tips, x="tip", bins=20)
plt.show()
```

```python
sns.histplot(data=tips, x="tip", bins=20, cumulative=True)
plt.show()
```

```python
sns.histplot(data=tips, x="tip", bins=20, hue="sex")
plt.show()
```

```python
sns.histplot(data=tips, x="tip", bins=20, hue="sex", stat="density")
# stat : {count (default), frequency, density, probability, percent, percentile}
plt.show()
```

```python
sns.histplot(data=tips, x="tip", bins=20, hue="sex", stat="count", multiple="dodge")
# multiple : {layer (default), dodge, stack, fill}
plt.show()
```

```python
sns.kdeplot(data=tips, x="tip", hue="sex", cumulative=False, fill=True)
plt.show()
```

```python
sns.ecdfplot(data=tips, x="tip", hue="sex")
plt.show()
```

```python
sns.displot(data=tips, x="tip", hue="sex", multiple="dodge")
plt.show()
```
