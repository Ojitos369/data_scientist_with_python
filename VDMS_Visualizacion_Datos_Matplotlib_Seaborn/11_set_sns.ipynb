```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.barplot(x=["A", "B", "C"], y=[1, 3, 2])
plt.show()
```

```python
sns.set(style="darkgrid")
sns.barplot(x=["A", "B", "C"], y=[1, 3, 2])
plt.show()
```

```python
sns.set(style="dark")
sns.barplot(x=["A", "B", "C"], y=[1, 3, 2])
plt.show()
```

```python
sns.set(style="darkgrid", palette="dark", font_scale=0.5)
sns.barplot(x=["A", "B", "C"], y=[1, 3, 2])
plt.show()
```

```python
sns.set_theme(style="darkgrid", palette="dark", font_scale=0.5)
sns.barplot(x=["A", "B", "C"], y=[1, 3, 2])
plt.show()
```
