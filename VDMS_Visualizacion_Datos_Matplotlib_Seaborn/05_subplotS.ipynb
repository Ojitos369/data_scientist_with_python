```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
# print(type(axes))
axes.plot(x, y, "r")
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, "r")
axes[1].plot(y, x, "g")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, y, "r")
ax2.plot(y, x, "b")
plt.show()
```

```python
fig, axes = plt.subplots(2, 4)
axes[0, 0].plot(x, np.cos(x), "r")
axes[0, 1].plot(x, np.sin(x), "g")
axes[0, 2].plot(x, np.tan(x), "b")
axes[0, 3].plot(x, np.cos(x)**2, "y")
# axes[1].plot(y, x, "g")
plt.show()
```

```python
fig, ((ax1,ax2,ax3,ax4), (ax5,ax6,ax7,ax8)) = plt.subplots(2, 4)
ax1.plot(x, np.cos(x), "r")
ax2.plot(x, np.sin(x), "g")
ax6.plot(x, np.tan(x), "b")
ax8.plot(x, np.cos(x)**2, "y")
fig.tight_layout()
plt.show()
```
