```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1, 50, 100)
```

```python
plt.style.use('dark_background')
# plt.style.use('default')
```

```python
plt.hist(data, color="purple", bins=20, width=2)
plt.show()
```

```python
plt.hist(data, color="blue", bins=20, histtype="step")
plt.show()
```

```python
plt.boxplot(data)

plt.show()
```

```python
plt.boxplot(data, patch_artist=True, tick_labels=["Label see"], zorder=5)

plt.show()
```
