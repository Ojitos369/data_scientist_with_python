```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
# tips.select_dtypes(include="number").corr()
corr = tips.corr(numeric_only=True)
corr
```

```python
sns.heatmap(corr)
plt.show()
```

```python
sns.heatmap(
    corr, 
    annot=True, 
    cmap='coolwarm',
    linewidths=5,
    linecolor='black',
    # vmin=-1,
    # vmax=1,
    cbar=True
)
plt.show()
```
