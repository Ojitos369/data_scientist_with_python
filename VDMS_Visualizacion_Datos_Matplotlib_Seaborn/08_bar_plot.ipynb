```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('dark_background')
# plt.style.use('default')
```

```python
country = ["India", "Japon", "Mexico", "Colombia", "Alemania"]
population = [1080, 808, 900, 100, 380]
```

```python
plt.figure(figsize=(5, 3))
plt.bar(country, population, width=0.5, color=["#0a8", "#81c", "#1a5"], label=country)
plt.xticks(np.arange(len(country)), ["IND", "JPN", "MEX", "COL", "ALE"], rotation=45)

plt.title("Poblacion por pais")
plt.xlabel("Pais")
plt.ylabel("Poplacion")
plt.legend(title="Paises")

plt.show()
```

```python
plt.barh(country, population, color=["#0a8", "#81c", "#1a5"], height=0.5)
plt.show()
```
