```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x, y, 'bo-')
plt.show()
```

```python
plt.plot(x, y, 'r*--')
plt.show()
```

```python
plt.plot(x, y, 'gD:')
plt.show()
```

```python
plt.hist(x, color="r", bins=20)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python

plt.scatter(x, y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```

```python
plt.stem(x, y, "r*:")
plt.show()
```

```python
plt.stackplot(x, y)
plt.show()
```

```python
plt.stairs(x, linestyle=":", color="r", linewidth=2)
plt.show()
```
