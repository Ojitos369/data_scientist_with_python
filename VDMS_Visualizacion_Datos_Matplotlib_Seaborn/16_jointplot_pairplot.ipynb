```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
sns.jointplot(
    data=tips, 
    x="total_bill", y="tip", 
    hue="sex", 
    kind="hist", 
    marginal_ticks=True,
    marginal_kws=dict(bins=25, fill=True, multiple="dodge"),
)
plt.show()
```

```python
sns.pairplot(
    data=tips,
    hue="sex",
    corner=True,
    kind="scatter",
)
# Pairplot cruza todas las variables numericas
plt.show()
```
