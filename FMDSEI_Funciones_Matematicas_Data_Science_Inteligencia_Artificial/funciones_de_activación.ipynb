```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

N = 100
x = np.linspace(-10, 10, num=N)
```

## Funci贸n lineal
$$y=mx+b$$

```python
def f(x):
    return x

y = f(x)

plt.plot(x, y)
plt.show()
```

## Funci贸n escal贸n o de Heaviside
$$H(x) =
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python

def f(x):
    h = np.zeros(len(x))
    for i, n in enumerate(x):
        h[i] = 0 if n < 0 else 1
    return h

y = f(x)

plt.plot(x, y)
plt.show()
```

## Funci贸n sigmoide

$$f(x)=\frac{1}{1-e^{-x}}$$

```python
def f(x):
    return 1/(1 + np.exp(-x))

y = f(x)

plt.plot(x, y)
plt.grid()
```

## Funci贸n tangente hiperb贸lica

$$f(x)=\frac{2}{1+e^{-2x}}-1$$

```python
def f(x):
    return np.tanh(x)

y = f(x)

plt.plot(x, y)
plt.grid()
```

## Funci贸n ReLU

$$R(x)=max(0,x)$$

```python
def f(x):
    return np.maximum(0, x)

y = f(x)

plt.plot(x, y)
plt.grid()
```

```python
# LeakyReLu

def f(x):
    t = np.zeros(len(x))
    for i, n in enumerate(x):
        t[i] = n if n > 0 else n * 0.01
    return t

y = f(x)

plt.plot(x, y)
plt.grid()
```
