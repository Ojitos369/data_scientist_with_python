```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))

categories = ["PA", "PB", "PC"]
sales = np.random.randint(90, 150, len(categories))

plt.bar(categories, sales, color=['red', 'green', 'blue'], label=categories)

plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.legend(title="Categories") # show the label
plt.show()
```
