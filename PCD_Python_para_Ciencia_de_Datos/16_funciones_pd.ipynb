```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df.info()
```

```python
p(df.isnull().sum())
p()
p(df.isna().sum())
```

```python
null_cid_list = df["CustomerID"].isna()
null_cid = df[null_cid_list]
null_cid
```

```python
replace = df["CustomerID"].fillna(df["CustomerID"].mean())
```

```python
df["replaced"] = replace
replaced_cid = df[null_cid_list]
replaced_cid
```

```python
df["finalPrice"] = (df["UnitPrice"] * 0.9) * df["Quantity"]
df
```

```python
df["dicount"] = df["UnitPrice"].apply(lambda x: x * 0.1)
df
```

```python
df["totalDiscount"] = df.apply(lambda x: (x["UnitPrice"] * 0.1) * x["Quantity"], axis=1)
df
```
