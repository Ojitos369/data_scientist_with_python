```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
arr = np.random.normal(170, 20, 200)
```

```python


# plt.hist(arr, color="salmon", edgecolor="black", alpha=0.6, bins=[150, 155, 160, 170, 180, 185, 190])
# plt.hist(arr, color="salmon", edgecolor="black", alpha=0.6, bins=15, range=(160, 180))
plt.hist(arr, color="salmon", edgecolor="black", alpha=0.6, bins=15, density=True)
plt.title("Distribucion de alturas")

plt.xlabel("altura")
plt.ylabel("cantidad")

plt.show()
```

```python
plt.figure(figsize=(8, 6))
np.random.seed(369)
ar2 = np.random.normal(30, 15, 50)
ar3 = np.random.normal(40, 15, 50)
ar4 = np.random.normal(35, 15, 50)
ages = ar2, ar3, ar4
labels = ["grupo1", "grupo2", "grupo3"]
plt.boxplot(ages, patch_artist=True, vert=True, notch=True, tick_labels=labels)
plt.title("Distribucion de edades")
plt.xlabel("grupo")
plt.ylabel("edad")
plt.show()
```
