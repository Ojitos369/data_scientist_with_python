```python
import numpy as np
from matplotlib import pyplot as plt, gridspec
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
data = np.random.randn(100)
```

```python
gs = gridspec.GridSpec(2, 2, height_ratios=[2, 1], width_ratios=[1, 1])
fig = plt.figure(figsize=(10, 8))

ax1 = fig.add_subplot(gs[0, :])
ax1.plot(x, y, color="b")
ax1.set_title("Seno de x")
ax1.set_xlabel("x")
ax1.set_ylabel("sen(x)")


ax2 = fig.add_subplot(gs[1, 1])
ax2.hist(data, bins=20, color="g")
ax2.set_title("Histograma")
ax2.set_xlabel("x")
ax2.set_ylabel("Frecuencia")


ax3 = fig.add_subplot(gs[1, 0])
ax3.plot(x, y, color="r", linestyle=":")
ax3.bar(x, y, color="yellow", width=0.1)
ax3.scatter(x, y, color="black")
ax3.set_title("Seno de x")
ax3.set_xlabel("x")
ax3.set_ylabel("sen(x) y Frecuencia")


plt.tight_layout()
plt.show()
```
