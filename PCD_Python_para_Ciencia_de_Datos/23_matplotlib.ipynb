```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
month = np.array(["E", "F", "M", "A", "Ma"])
sales = np.array([20, 25, 38, 28, 35])
```

```python
plt.figure(figsize=(8, 6))
plt.plot(month, sales, marker="o", color="blue")

plt.title("Ventas Mensuales de un producto")
plt.xlabel("Meses")
plt.ylabel("Ventas en miles de unidades")

plt.show()
```

```python
plt.figure(figsize=(8, 6))

hours = [2, 3, 4, 5, 6, 7, 8]
exam = [55, 60, 65, 70, 75, 80, 85]

plt.scatter(hours, exam, color="green", marker="o")
plt.title("Relacion entre estudio y puntaje")
plt.xlabel("Horas")
plt.ylabel("Puntaje")
plt.show()
```

```python
plt.figure(figsize=(8, 6))

hours = [2, 3, 4, 5, 6, 7, 8]
exam_1 = [55, 60, 65, 70, 75, 80, 85]
exam_2 = [60, 55, 88, 50, 66, 75, 90]
exam_3 = np.random.randint(50, 100, 7)
plt.plot(hours, exam_3, color="purple", marker='^', label="Estudiante 3", linestyle="--")
# plt.scatter(hours, exam_1, color="green", marker="o", label="Estudiante 1", linestyle="--")
# plt.scatter(hours, exam_2, color="blue", marker="x", label="Estudiante 2", linestyle="-.")
plt.scatter(hours, exam_1, color="green", marker='4', label="Estudiante 1")
plt.scatter(hours, exam_2, color="red", marker=4, label="Estudiante 2")
plt.title("Relacion entre estudio y puntaje")
plt.xlabel("Horas")
plt.ylabel("Puntaje")
plt.show()
```

```python
# plt.figure(figsize=(8, 6))
exam_3 = np.random.randint(50, 100, 7)
exam_4 = np.random.randint(50, 100, 7)
plt.plot(hours, exam_3, color="green", marker='^', label="Estudiante 3", linestyle="--")
plt.plot(hours, exam_4, color="blue", marker=8, label="Estudiante 4", linestyle="-.")
plt.title("Relacion entre estudio y puntaje")
plt.xlabel("Horas")
plt.ylabel("Puntaje")
plt.show()
```
