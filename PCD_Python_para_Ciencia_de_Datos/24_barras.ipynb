```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))

categories = ["PA", "PB", "PC"]
sales = np.random.randint(90, 150, len(categories))

plt.bar(categories, sales, color=['red', 'green', 'blue'], label=categories)

plt.annotate(
    "Maximo de las ventas", 
    xy=(categories[sales.argmax()], sales.max()),
    arrowprops=dict(facecolor='black', shrink=0.05),
    xytext=(categories[sales.argmax()], sales.max()//2),
)
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.legend(title="Categories") # show the label
plt.show()
```

```python
plt.figure(figsize=(8, 6))

categories = ["PA", "PB", "PC"]
sales = np.random.randint(90, 150, len(categories))

plt.barh(categories, sales, color=['red', 'green', 'blue'], label=categories)

plt.title('Sales by Category')
plt.ylabel('Category')
plt.xlabel('Sales')
plt.show()
```
