```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))

products = ["PA", "¨PB", "PC"]
sales = np.random.randint(10, 50, 3)
hatch=['**O', 'oO', 'O.O', '.||.']

plt.pie(sales, labels=products, autopct="%1.2f%%", shadow=True, startangle=90, hatch=hatch)

plt.title("Sales by product")
plt.show()
```
