```python
import pandas as pd
```

```python
df = pd.read_csv('./files/athletes_new.csv')
df.head(2)
```

```python
df.shape[0]
```

```python
fem = df["gender"] == "Female"
# dfm = df[fem] -> solo mujeres
dfm = df[-fem] # -> solo hombres, el "-" niega la condicion
dfm.head(2)
```

```python
nmx = df["country_code"] != "MEX"

hnm = df[-fem & nmx] # -> hombres que no son de mexico
hnm.shape[0]
```

```python
hnm.head(2)
```

```python
mmf = df[fem & -nmx] # -> mujeres que son de mexico
mmf.shape[0]
```
