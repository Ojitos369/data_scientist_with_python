```python
d = {
    "Key": ["K0", "K1", "K2", "K3"],
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"],
}
der = pd.DataFrame(d)
der
```

### Merge

```python
izq.merge(der)
# izquierda es el de afuera, derecha es el de adentro
```

```python
d = {
    "key": ["K0", "K1", "K2", "K3"],
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"],
}
izq2 = pd.DataFrame(d)
d = {
    "key2": ["K0", "K1", "K2", "K3"],
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"],
}
der2 = pd.DataFrame(d)
```

```python
izq2.merge(der2, on="key")
# Esto falla porque der2 no tiene key
```

```python
izq2.merge(der2, left_on="key", right_on="key2")
# se especifica con cuales se unen cada columna
```

```python
d = {
    "key": ["K0", "K1", "K2", "K3"],
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"],
}
izq3 = pd.DataFrame(d)
d = {
    "key2": ["K0", "K1", "K2", np.nan],
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"],
}
der3 = pd.DataFrame(d)
```

```python
izq3.merge(der3, left_on="key", right_on="key2")
# no trae los nulos y deja los que coinciden
```

```python
izq3.merge(der3, left_on="key", right_on="key2", how="left")
# how, inner por default, left, right, outer
```

```python
izq3.merge(der3, left_on="key", right_on="key2", how="right")
```

```python
izq3.merge(der3, left_on="key", right_on="key2", how="outer")
```
