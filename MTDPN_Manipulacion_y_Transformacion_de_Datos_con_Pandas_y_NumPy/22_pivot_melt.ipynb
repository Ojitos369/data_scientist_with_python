```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./files/bestsellers-with-categories.csv')
```

```python
df.head()
```

### Pivot Table

```python
df.pivot_table(index='Author',columns='Genre',values='User Rating')
```

```python
df.pivot_table(index='Author',columns='Year',values='Reviews')
```

```python
df.pivot_table(index='Genre',columns='Year', values='User Rating',aggfunc='mean')
```
